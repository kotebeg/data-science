```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

from scipy.stats import boxcox
```

```python
df = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
df.head()
```

```python
# difference between different transformations
fig = px.line(df, height=500, width=900)
fig.update_traces(mode="lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
df.head()
```

```python
fig = px.line(df,y = 'SqrtPassengers', height=500, width=900)
fig.update_traces(mode="lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
df['LogPassengers'] = np.log(df['Passengers'])
df.head(3)
```

```python
fig = px.line(df, height=500, width=900)
fig.update_traces(mode="lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
# SqrtPassengers vs LogPassengers
fig = px.line(df, y = ['SqrtPassengers', 'LogPassengers'],height=500, width=900)
fig.update_traces(mode="lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")
fig.show()
```

```python
# SqrtPassengers vs LogPassengers
fig = px.line(df, y = ['LogPassengers'],height=500, width=900)
fig.update_traces(mode="lines", hovertemplate=None)
fig.update_layout(hovermode="x unified")
fig.show()
```
