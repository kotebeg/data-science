```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
df.head()
```

```python
df['Passengers'].plot(figsize=(10, 4));
```

```python
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
df['SqrtPassengers'].plot(figsize=(10, 4));
```

```python
df['LogPassengers'] = np.log(df['Passengers'])
df['LogPassengers'].plot(figsize=(10, 4));
```

```python
df.head()
```

```python
df[['LogPassengers', 'SqrtPassengers']].plot(figsize=(10, 4));
```

```python
data, lam = boxcox(df['Passengers'])
lam
```

```python
df['BoxCoxPassengers'] = data
df['BoxCoxPassengers'].plot(figsize=(10, 4));
```

```python
df.head()
```

```python
df[['SqrtPassengers', 'LogPassengers', 'BoxCoxPassengers']].plot(figsize=(10, 4));
```

```python
df[['LogPassengers', 'BoxCoxPassengers']].plot(figsize=(10, 4));
```
